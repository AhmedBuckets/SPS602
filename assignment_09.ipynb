## **Bar Plot**

A barplot shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.

# My Bar Plot Recreations

```python
# The x axis will be specified to have smoker information while the y axis will have the total bill information
# If we want each category in the "smoker" column to have its own color we can use color = "smoker"
# If we want to further break down each bar using information in another column we can use facet_col = "sex"
fig = px.bar(df, x="smoker", y="total_bill", color="smoker", facet_col="sex")
fig.show()
```

```python
# For this one we are counting the number of days, so that will be the x-axis
# We want a different color for each day and so we can use color = day
# To split them out by gender we can sue facet_col = sex
fig = px.bar(df, x="day", y="total_bill", color="day", facet_col="sex")
fig.show()
```

# My Scatter Recreations

```python
# This figure will plot total bill amount on the x axis against tip on the y axis
# To have a separate color per category in "sex" we can use color = "sex"
# To split the data out to have a separate scatter per smoker category we can use facet_col = smoker
fig = px.scatter(df, x="total_bill", y="tip", color="sex", facet_col="smoker")

fig.show()
```

```python
# The trendline tag lets us add a trendline per scatter per sex
fig = px.scatter(df, x="total_bill", y="tip", color="sex", facet_col="smoker", trendline="ols")

fig.show()
```

```python
# To break out the scatters so that every column represents a day we can use facet_col = "day"
# If we have facet_row = "time", each row will represent a category in the column "time", thus the two rows Dinner and Lunch below
fig = px.scatter(df, x="total_bill", y="tip", facet_row = "time", facet_col="day")

fig.show()
```

# **Histogram**

A histogram takes as input a numeric variable only. The variable is cut into several bins, and the number of observation per bin is represented by the height of the bar. It is possible to represent the distribution of several variable on the same axis using this technique.

```python
#Exploring distrubution of tip
```

# My Histogram Recreation

```python
fig = px.histogram(df, x="tip", nbins=20)
fig.update_xaxes(tickmode="linear", dtick=2)

fig.show()
```

# **Boxplot**

A boxplot gives a nice summary of one or several numeric variables. The line that divides the box into 2 parts represents the median of the data

# My Boxplot Recreation

```python
fig = px.box(df, x="smoker", y="total_bill", color="smoker")
fig.show()
```
